```python
import matplotlib.pyplot as plt 
x = dataset['OURAREA']
y = dataset['OURPRICE']
plt.plot(x,color='orange',label='OURAREA')
plt.plot(y,color='g',label='OURPRICE')
plt.xlabel('OURAREA')
plt.xlabel('OURPRICE')
plt.title('HOUSING PRICES')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = dataset['OURAREA']
y = dataset['OURPRICE']
plt.plot(x,color='b',marker='o',markersize=12,label='OURAREA')
plt.plot(y,color='r',linestyle='dashed',linewidth=2,label='OURPRICE')
plt.xlabel('OURAREA')
plt.xlabel('OURPRICE')
plt.title('HOUSING PRICES')
plt.legend()
plt.show()
```
